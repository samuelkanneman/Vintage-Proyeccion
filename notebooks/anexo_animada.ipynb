## Anexo: Curvas Vintage Animadas

Animación que muestra la evolución del índice de morosidad cohorte por cohorte.
Cada cohorte se dibuja MOB a MOB, y al completarse queda como rastro semitransparente
mientras la siguiente cohorte comienza a animarse.

La animación se genera ejecutando `src/generar_animacion_vintage.py` y se guarda como GIF en `reports/curvas_vintage_animado.gif`.

```python
result = subprocess.run(
    ["py", os.path.join(SRC_DIR, "generar_animacion_vintage.py")],
    capture_output=True, text=True, cwd=PROJECT_ROOT
)
print(result.stdout)
if result.returncode != 0:
    print("STDERR:", result.stderr)

# Mostrar el GIF generado
from IPython.display import HTML, display
gif_path = os.path.join(PROJECT_ROOT, "reports", "curvas_vintage_animado.gif")
if os.path.isfile(gif_path):
    display(HTML(f'<img src="../reports/curvas_vintage_animado.gif" width="900">'))
else:
    print("No se encontró el GIF. Revise los errores anteriores.")
```
